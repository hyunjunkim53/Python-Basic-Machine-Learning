## 1. Demographic Filtering (인구통계학적 필터링)

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('tmdb_5000_credits.csv')
df2 = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df1.columns = ['id', 'title', 'cast', 'crew'] # tmdb_5000_credits.csv 파일의 컬럼명 변경
df1.columns
```

```python
df1[['id', 'cast', 'crew']] # title값이 df2와 동일하므로 데이터 중복 방지 위해 제거
```

```python
df2 = df2.merge(df1[['id', 'cast', 'crew']], on='id') # id를 기준으로 df2 뒤에 df1 병합
df2.head(3)
```

영화 1 : 영화의 평점이 10/10 -> 5명이 평가

영화 2 : 영화의 평점이 8/10 -> 500명이 평가 # 신뢰도 더 높음

```python
C = df2['vote_average'].mean() # C = 전체 영화의 평균 평점
C
```

```python
m = df2['vote_count'].quantile(0.9) # m = 상위 10%에 해당하는 평가수를 가지는 영화의 기준
m
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m] # q_movies = 평점 개수가 1838개 이상인 데이터
q_movies.shape
```

```python
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C) # 제공한 공식 적용
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1) # q_movies가 가지는 모든 데이터들에 대해, 각 행을 weighted_rating 함수를 통과 시키고, 해당 값을 score의 값으로 저장
q_movies.head(3)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
pop= df2.sort_values('popularity', ascending=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

plt.barh(pop['title'].head(10),pop['popularity'].head(10), align='center',
        color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular Movies")
```

1. TfidfVectorizer (TF-IDF 기반의 벡터화): 단순히 많이 등장한다고 해서 중요한 단어라고 판단하지 않고, 해당 문서에서만 특별하게 많이 등장하는 단어에 더 높은 가중치를 주는 방식
2. CountVectorizer (빈도 기반 벡터화): 각 문서에서 단어가 나타나는 횟수를 그대로 숫자로 변환하는 방식 (단점: 문장에서 의미는 없지만 자주 등장하는 단어들이 높은 값을 갖게 되어, 정작 중요한 핵심 단어를 파악하기 어려울 수 있음)


```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english') # stop_words='english': 불용어 제거
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
ENGLISH_STOP_WORDS
```

```python
df2['overview'].isnull().values.any()
```

```python
df2['overview'] = df2['overview'].fillna('')
```

```python
tfidf_matrix = tfidf.fit_transform(df2['overview'])
tfidf_matrix.shape
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix) # 영화 줄거리(overview) TF-IDF 벡터 간의 내적을 통해 코사인 유사도 점수 행렬(4803x4803) 생성
cosine_sim
```

```python
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
indices
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index 값을 얻기
    idx = indices[title]
    
    # 코사인 유사도 매트릭스 (cosine_sim) 에서 idx 에 해당하는 데이터를 (idx, 유사도) 형태로 얻기
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 코사인 유사도 기준으로 내림차순 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 자기 자신을 제외한 10개의 추천 영화를 슬라이싱
    sim_scores = sim_scores[1:11]
    
    # 추천 영화 목록 10개의 인덱스 정보 추출
    movie_indices = [i[0] for i in sim_scores]
    
    # 인덱스 정보를 통해 영화 제목 추출
    return df2['title'].iloc[movie_indices]
```

```python
# 감독 정보를 추출
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df2['director'] = df2['crew'].apply(get_director)
df2['director']
```

```python
# 처음 3개의 데이터 중에서 name 에 해당하는 value 만 추출
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ', ''))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2
```
